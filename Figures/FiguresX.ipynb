```python
import numpy as np
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    "SURVAVIBILITE": 10,
    "CONTRLER": 9,
    "FIABILIT": 5,
    "LOCALISER": 6,
    "ENTRETIEN": 2,
    "LONGVIT": 4,
    "UTILISATION": 8,
    "ALIMENTATION DES SYSTMES": 9,
    "COMPATIBILIT DES SYSTMES": 9,
}

data = {k: v for k, v in sorted(data.items(), key=lambda item: item[1], reverse=True)}
besoins = list(data.keys())
importance = list(data.values())

fig, ax = plt.subplots(figsize=(12, 6))

# Creating the bar plot
bars = ax.bar(besoins, importance, color="maroon", width=0.4)

# Correcting the xlabel and ylabel functions
ax.set_ylabel("Importance", fontsize=15)
ax.set_xlabel("Besoins", fontsize=17)

added = []
for bar, val in zip(bars, importance):
    if val not in added:
        added.append(val)

        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.1,
            str(val),
            ha="center",
            va="bottom",
        )

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
# Adding rotation to x-axis labels
ax.set_yticks([])
ax.set_yticklabels([],fontsize = 14)
ax.set_xticklabels(besoins, rotation=30, ha="right")
ax.spines["left"].set_visible(False)


plt.show()
```
